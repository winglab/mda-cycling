```python
import requests
import pandas as pd
import warnings
from datetime import datetime

warnings.filterwarnings('ignore')
```

```python

## Function to extract CSV files from the website
def extract_and_download_data():

    ## Assign the base URLs
    base_url = "https://opendata.apps.mow.vlaanderen.be/fietstellingen/"

    ## Generate URLs programmatically to bypass the JavaScript-rendered HTML
    dates = pd.date_range(start = '2019-08-01', end = datetime.today(), freq = 'MS')

    ## Define columns (sample ones) explicitly since the CSV lacks a header row
    col_names = ["Site_ID", "Direction", "Modus", "Start_Time", "End_Time", "Count"]
    dataframes = []

    for date in dates:
        file_name = f"data-{date.strftime('%Y-%m')}.csv"
        file_url = base_url + file_name

        ## Ping the server to see if this specific month's file exists
        response = requests.head(file_url)
        if response.status_code == 200:
            print(f"Fetching {file_name}...")
            try:
                df = pd.read_csv(file_url, sep = ',', header = None, names = col_names, low_memory = False)

                ## Enforce explicit grammar tags over numerical identifiers for categorical data
                df['Site_ID'] = df['Site_ID'].astype(str).apply(lambda x: f"Location_Tag_{x}")

                dataframes.append(df)
            except Exception as e:
                print(f"Failed to parse {file_name}: {e}")

    if dataframes:
        master_df = pd.concat(dataframes, ignore_index=True)
        print(f"Success! Master dataset compiled with {len(master_df)} rows.")
        return master_df
    else:
        print("No files could be downloaded.")
        return None

## Execute the historical pipeline
full_dataset = extract_and_download_data()
```

```python
## Download the CSV file
full_dataset.to_csv('Full_Dataset_Cycling.csv', index = False)
```
